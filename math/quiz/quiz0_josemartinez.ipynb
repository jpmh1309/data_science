```python
# Quiz # 0 
# Student: Jose Martinez Hdez
# Course: Data Science 
# Year: 2021

# Import libraries
import random
import numpy as np
import matplotlib.pyplot as plt

# Rectified Linear Unit Function
relu = lambda x: np.maximum(0.0, x)

# Smooth Approximation of Rectified Linear Unit Function
softplus = lambda x: x * (x >= 0) + np.log1p(np.exp(-np.abs(x)))

# Function to check linearity F[a*x+b*y] = a*F[x] + b*F[y]
def check_linearity(function, x, y, a, b):
  LHS = function(a*x+ b*y)
  RHS = a*function(x) + b*function(y)
  tolerence = 1e-4
  if np.sum(LHS-RHS) < tolerence:
    print('Linear System')
    return True
  else:
    print('Not a Linear System')
    return False
```

To demonstrate whether a function is linear or not, it must satisfy with the following property:

$$ H(\alpha f_1(x) + \beta f_2(x)) = \alpha H(f_1(x)) + \beta H(f_2(x))$$

Since the function function Relu isn´t linear for all values of $ x, y, a, b$, this function is not strictly linear, but it could be classified as piece-wise linear because for $ x, y, a, b > 0$ it satisfies the property of superposition.
