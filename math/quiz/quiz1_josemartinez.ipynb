```python
# Quiz # 1 
# Student: Jose Martinez Hdez
# Course: Data Science 
# Year: 2021

import torch

def matrix_propierty_1(A):
  # Sanity check 
  if len(A) != len(A[0]):
    print("Matrix isn´t nxn")
    return False
  if torch.det(A) == torch.tensor(0):
    print("Matrix isn´t invertible")
    return False

  # Local variables 
  A_inv = torch.inverse(A)
  A_t = torch.transpose(A, 0, 1)
  A_inv_t = torch.transpose(A_inv, 0, 1)
  
  # Calculate left side 
  left = torch.mm(A_inv_t, A_t)
  print("-> left side:\n{}".format(left))

  # Calculate right side 
  right = torch.mm(A_t, A_inv_t)
  print("-> right side:\n{}".format(right))

  # Calculate and print ¨left-right¨ 
  M = torch.add(left, -right)
  print("-> left side - right side:\n{}".format(M))

  # Compare all values of the matrix and take into account 
  # small precision differences
  return torch.all(torch.lt(torch.abs(M), 1e-6))

# Create random matrix 
A = torch.randn(3,3)

# Demostrate whether the equation is true or not 
print("Is the equation true? \n- Answer: {}".format(matrix_propierty_1(A)))
```

Se tiene que:

$$\left(A^{-1}\right)^{T}A^{T}=A^{T}\left(A^{-1}\right)^{T}$$

  Donde $A\in\mathbb{R}^{n\times n}$ invertible. Resolviendo la ecuación anterior del lado izquierdo y por propiedades de la traspuesta se llega a:

$$\left(A^{-1}A\right)^{T}=A^{T}\left(A^{-1}\right)^{T}$$

Al $A$ ser una matriz invertible, cumple con la definición de:

$$A^{-1}A = I = AA^{-1}$$

Sustituyendo la ecuación anterior en el lado izquierdo, se llega a:

$$\left(AA^{-1}\right)^{T}=A^{T}\left(A^{-1}\right)^{T}$$

Aplicando las propiedades de la traspuesta del lado izquierdo se llega a:

$$A^{T}\left(A^{-1}\right)^{T} = A^{T}\left(A^{-1}\right)^{T}$$ 

```python
import torch

def matrix_propierty_2(A, B):
  # Sanity check
  if len(A) != len(A[0]):
    print("Matrix isn´t nxn")
    return False
  if len(A) != len(B):
    print("Matrix A and B aren't same size")
    return False

  # Calculate left side 
  left1 = torch.add(A, B)
  left2 = torch.add(A, -B)
  left = torch.mm(left1, left2)
  print("-> left side:\n{}".format(left))

  # Calculate right side 
  right1 = torch.mm(A, A)
  right2 = torch.mm(B, B)
  right = torch.add(right1, -right2)
  print("-> right side:\n{}".format(right))

  # Calculate and print ¨left-right¨ 
  M = torch.add(left, -right)
  print("-> left side - right side:\n{}".format(M))

  # Compare all values of the matrix and take into account 
  # small precision differences
  return torch.all(torch.lt(torch.abs(M), 1e-6))

# Create random matrix 
A = torch.randn(3,3)
B = torch.randn(3,3)

# Demostrate whether the equation is true or not 
print("Is the equation true? \n- Answer: {}".format(matrix_propierty_2(A, B)))
```
